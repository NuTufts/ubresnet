```python
image2d_wire_chain = ROOT.TChain("image2d_wire_tree")
image2d_labels_chain = ROOT.TChain("image2d_labels_tree")
image2d_weights_chain = ROOT.TChain("image2d_weights_tree")
image2d_background_chain = ROOT.TChain("image2d_background_tree")
image2d_track_chain = ROOT.TChain("image2d_track_tree")
image2d_shower_chain = ROOT.TChain("image2d_shower_tree")
image2d_track_end_chain = ROOT.TChain("image2d_track_end_tree")
image2d_full_image_chain = ROOT.TChain("image2d_full_image_tree")
```

```python
for _file in _files: image2d_wire_chain.AddFile(_file)
print ('Found', image2d_wire_chain.GetEntries(), 'entries in image2d wire')

for _file in _files: image2d_labels_chain.AddFile(_file)
print ('Found', image2d_labels_chain.GetEntries(), 'entries in image2d labels')

for _file in _files: image2d_weights_chain.AddFile(_file)
print ('Found', image2d_weights_chain.GetEntries(), 'entries in image2d weights')

for _file in _files: image2d_background_chain.AddFile(_file)
print ('Found', image2d_background_chain.GetEntries(), 'entries in image2d background')

for _file in _files: image2d_track_chain.AddFile(_file)
print ('Found', image2d_track_chain.GetEntries(), 'entries in image2d track')

for _file in _files: image2d_shower_chain.AddFile(_file)
print ('Found', image2d_shower_chain.GetEntries(), 'entries in image2d shower')

for _file in _files: image2d_track_end_chain.AddFile(_file)
print ('Found', image2d_track_end_chain.GetEntries(), 'entries in image2d track_end')

for _file in _files: image2d_full_image_chain.AddFile(_file)
print ('Found', image2d_full_image_chain.GetEntries(), 'entries in image2d full_image')
```
